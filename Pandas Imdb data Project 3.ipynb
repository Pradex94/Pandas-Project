# 10. In Which Year There Was The Highest Average Voting?

```python
# data[data.groupby('Year')['Votes'].mean().sort_values(ascending=False)][['Year']]
# Calculate the mean votes per year
mean_votes = data.groupby('Year')['Votes'].mean()
```

```python
# Sort the years by mean votes in descending order
sorted_years = mean_votes.sort_values(ascending=False)
```

```python
# Filter the original DataFrame to keep only the years with the highest average votes
result = data[data['Year'].isin(sorted_years.index)][['Year']]
result
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.espncricinfo.com/records/most-runs-in-career-282827'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')

# Find table and extract data
table = soup.find('table')
df = pd.read_html(str(table))[0]
print(df.head())
```
